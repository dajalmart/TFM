```python
import pandas as pd
import numpy as np
from math import factorial
import matplotlib.pyplot as plt
import networkx as nx
import seaborn as sns
```

```python
credit = pd.read_excel('data/2 parte 2007 2.0.xlsx', index_col=0)
credit.info()
credit.head()
```

```python
credito = credit.dropna()
credito
```

```python
Firms=np.transpose(credito.values.tolist())
nfirms=credito.shape[1]
#nfirms=85
```

```python
#vamos a construir la matriz de adyacenia 

A = np.zeros((nfirms,nfirms)) 
print(A)
for i in range (nfirms):
    for j in range (nfirms):
        ro=np.corrcoef(Firms[i],Firms[j])[0,1]
        d = (2*(1-ro))**0.5
        A[i,j]=d
        #al ser mat simétrica
        A[j,i] = d
A
```

```python
#definimos la medida de centralidad hasta tercer orden

# en la lista Re guardaremos cada valor de la medida de centralidad para cada e diferente de las 35 empresas

Re=[]
ones = np.ones(nfirms)
C=np.dot(A,A)
#D=A^4
D=np.dot(C,C)
for e in np.arange(0.01,1, 0.01):

    Rv=np.identity(nfirms)+A*e + (A*e)**2/factorial (2) + (A*e)**3/factorial (3)+(A*e)**4/factorial (4)

    #Rv=np.exp(e*A)
    Re.append(np.dot(Rv,ones))
  
```

```python
# R es una matriz donde cada componente es la medida de centralidad para un riesgo diferente de una misma empresa
R=np.transpose(Re)
# Para hacer el ranking necesitamos solo Re ya que se hace comparando la centralidad de dif empresas para un mismo e
print(len(Re))
Re[5]
```

```python
rank=[]
for i in range (len(Re)):
    test_list = Re[i]  

    res = sorted(range(len(test_list)), key = lambda sub: test_list[sub], reverse=True)
    # ahora que están ordenados de mayor a menor, ponemos el ranking de cada uno
    b = np.zeros(nfirms) 

    b[res] = np.arange(nfirms) # numpy permite estas cosas
    rank.append(b)
```

```python
rankingboxplot=[]
for i in range (nfirms):
    rankingboxplot.append(ranking[i])
    
# Creating plot
plt.rcParams['font.size'] = 10
fig = plt.figure(figsize =(20, 10))
plt.boxplot(rankingboxplot)
plt.xlabel('Asset number') 
plt.ylabel('Ranking position') 
# show plot
plt.show()
```
