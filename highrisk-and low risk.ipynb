```python
import pandas as pd
import numpy as np
from math import factorial
import matplotlib.pyplot as plt
import networkx as nx
import seaborn as sns
```

```python
nfirms=100
e=1
ones = np.ones(nfirms)
Rp=[]
for p in np.arange(0.1,0.96,0.05):
    Rn=[]
    for n in range (1000):
        G = nx.erdos_renyi_graph(nfirms,p)
        A = nx.to_numpy_array(G)
        Rv=np.identity(nfirms)+A*e + (A*e)**2/factorial (2) + (A*e)**3/factorial (3)+(A*e)**4/factorial (4)
        Rn.append(np.dot(Rv,ones)[1]/np.mean(np.dot(Rv,ones)))
    Rp.append(Rn)
```

```python
labels = [0.1,0.15,0.2,0.25,0.3,0.35,0.4,0.45,0.5,0.55,0.6,0.65,0.7,0.75,0.8,0.85,0.9,0.95]
Rpbox=[]
p=np.arange(0.1,0.96,0.05)
for i in range (len(p)):
    Rpbox.append(Rp[i])   
# Creating plot
plt.rcParams['font.size'] = 13
fig = plt.figure(figsize =(20, 10))
plt.boxplot(Rpbox,labels=labels)
plt.xlabel('Density') 
plt.ylabel('$R_i$ / E($R_i$)') 
# show plot
plt.show()
```
