```python
import networkx as nx
```

```python
# Barabasi-Albert graph of 1000 nodes and m=2 links per new node.
ba = nx.barabasi_albert_graph(1000, 2)

# degrees of nodes
degrees = list(dict(ba.degree()).values())

# plot degree distribution
import matplotlib.pyplot as plt
import numpy as np

bins = np.geomspace(min(degrees), max(degrees), 21)
y, x = np.histogram(degrees, bins=bins)
plt.loglog(x[:-1], y/sum(y), 'o', label='empirical')
plt.xlabel('Degree')
plt.ylabel('Probability')

# plot theoretical distribution
from scipy.special import zeta

plt.loglog(x, x**(-2)/(zeta(2)-1), label='theoretical')
plt.legend()
```

```python
from scipy.special import factorial
er = nx.erdos_renyi_graph(1000, 0.01)

# degrees of nodes
degrees = list(dict(er.degree()).values())

# plot degree distribution
import matplotlib.pyplot as plt
import numpy as np

bins = np.geomspace(min(degrees), max(degrees), 21)
y, x = np.histogram(degrees, bins=bins)
plt.loglog(x[:-1], y/sum(y), 'o', label='empirical')
plt.xlabel('Degree')
plt.ylabel('Probability')

# plot theoretical distribution
from scipy.special import zeta

  
plt.loglog(x, np.exp(-10)*np.power(10, x)/factorial(x), label='theoretical')
plt.legend()
```
